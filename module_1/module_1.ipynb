```python
#создадим функцию для преобразования и подсчёта данных, разделённых символом "|"
def splitter(series):
    list = []
    for string in series:
        for item in string.split('|'):
            list.append(item)
    return Counter(list)
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
data27i=data.copy()
pairs = Counter()
for i in range(0,len(data27i)):
    artists = data27i.cast[i].split('|')
    for j in list(combinations(artists,2)):
        if j not in pairs:
            pairs[j]=1
        else:
            pairs[j]+=1
pairs.most_common(10)
```
